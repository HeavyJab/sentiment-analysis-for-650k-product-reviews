# Sentiment Feature
Using the AFINN we can obtain a sentiment score from -5 to +5 for each token. We can compute the total score and append it at the end of the DTM as a feature. 

```r
# put into a data frame format
df <- tibble(line=1:650000, text=train$text)

# remove puctuations and digits
df <- df %>% 
 mutate(text = str_replace_all(text, "[[:punct:]]|[[:digit:]]", "")) 

# unnest_tokens function to tokenize the frame
tokens <- df %>% 
  unnest_tokens(word, text) %>% 
  #remove the stopwords using anti_join()
  anti_join(stop_words)

# AFINN sentiment -5 - 5
sentiment <- get_sentiments("afinn")

# join the sentiment words with token
tokens <- tokens %>% 
  left_join(sentiment)
```
